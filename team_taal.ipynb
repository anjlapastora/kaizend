```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
fn = "../input/wikipedia_edits.csv.zip"
df = pd.read_csv(fn)
```

```python
df_anon_users = df[df['userid']==0] #for anonumous users, userid=0
df_anon_users
```

```python
# Show the number of maximum edits of anonymous users

df_anon_users['user'].value_counts()
```

```python
# 2: Artists with the most registered users
artists_reg_users = df[df['userid']!=0] 
artists_reg_users
artists_reg_users['title'].value_counts()
```

```python
# Most number of unregistered users/editors per title
df_anon_users['title'].value_counts()
```

```python
# 3 Per user, count how many careers
# get the user 
unregistered_users = df[df['userid']==0]
users = pd.DataFrame(unregistered_users.groupby(["user", "career"])["career"].agg([len])).reset_index()
users1 = users.groupby("user")["career"].agg([len]).sort_values(by="len", ascending=False)
users1
```

```python
registered_users = df[df['userid']!=0]
users = pd.DataFrame(registered_users.groupby(["user", "career"])["career"].agg([len])).reset_index()
users1 = users.groupby("user")["career"].agg([len]).sort_values(by="len", ascending=False)
users1
```
